# Explore Views


```python
%sql
USE CATALOG dae_learn_path;

CREATE TABLE IF NOT EXISTS smartphones
(id INT, name STRING, brand STRING, year INT);

INSERT INTO smartphones
VALUES (1, 'iPhone 14', 'Apple', 2022),
      (2, 'iPhone 13', 'Apple', 2021),
      (3, 'iPhone 6', 'Apple', 2014),
      (4, 'iPad Air', 'Apple', 2013),
      (5, 'Galaxy S22', 'Samsung', 2022),
      (6, 'Galaxy Z Fold', 'Samsung', 2022),
      (7, 'Galaxy S9', 'Samsung', 2016),
      (8, '12 Pro', 'Xiaomi', 2022),
      (9, 'Redmi 11T Pro', 'Xiaomi', 2022),
      (10, 'Redmi Note 11', 'Xiaomi', 2021)
```

```python
%sql
SHOW TABLES
```

```python
%sql
CREATE VIEW vw_apple_phones
AS SELECT * 
  FROM smartphones
  WHERE brand = 'Apple'
```

```python
%sql
SELECT * FROM vw_apple_phones
```

```python
%sql
CREATE TEMP VIEW tp_vw_phones_brands
AS SELECT DISTINCT brand 
  FROM smartphones;
  
SELECT * FROM tp_vw_phones_brands;
```

```python
%sql
SHOW TABLES
```

```python
%sql
CREATE GLOBAL TEMP VIEW gb_vw_latest_phones
AS SELECT *
  FROM smartphones
  WHERE year > 2020
  ORDER BY year DESC;
  
-- [NOT_SUPPORTED_WITH_SERVERLESS] GLOBAL TEMPORARY VIEW is not supported on serverless compute. SQLSTATE: 0A000
SELECT * FROM gb_vw_latest_phones;
```
